# US Net International Migration

Combines Census Bureau Population Estimates for two decades (2001 to 2010 from the 2010 evaluation estimates and 2011 to 2019 from the 2019 vintage population estimates), pulls out international migration data for states, sums the data for the US, and plots the past 20 years.

```python
%matplotlib inline
import pandas as pd
import os
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
plt.style.use('seaborn-whitegrid')
import pandas.io.formats.format as pf

# display numbers (floats) with thousand separator
pd.options.display.float_format = '{:,.1f}'.format

# found this hack to format integers for display with thousand separator
# https://stackoverflow.com/questions/29663252/format-pandas-integers-for-display?answertab=active#tab-top
class IntArrayFormatter(pf.GenericArrayFormatter):
    def _format_strings(self):
        formatter = self.formatter or '{:,d}'.format
        fmt_values = [formatter(x) for x in self.values]
        return fmt_values
pf.IntArrayFormatter = IntArrayFormatter
```

```python
#paths and variables
dir_path = os.getcwd()
data_path=os.path.join(dir_path,'source_data')
out_path=os.path.join(dir_path,'output')
file2010s='co-est2019-alldata.csv'
file2000s='co-est2010-alldata.csv'
```

## Data for the 2010s

```python
pep2010s=pd.read_csv(os.path.join(data_path,file2010s),converters={'STATE':str,'COUNTY':str},encoding='LATIN-1')
pep2010s.head()
```

```python
# Pull out records for states and columns for international migration
codes=['STATE','COUNTY']
names=[c for c in pep2010s.columns if c.endswith('NAME')]
net_int=[c for c in pep2010s.columns if c.startswith('INTERNATIONALMIG')]
keep_cols=codes+names+net_int
states2010s=pep2010s.loc[pep2010s['COUNTY']=='000',keep_cols]
states2010s.columns = states2010s.columns.str.replace('INTERNATIONALMIG','')
states2010s.head()
```

```python
states2010s.info()
```

```python
usa2010s=states2010s.sum(axis=0,numeric_only=True)
usa2010s
```

```python
# Drop 2010, as this measures migration between Apr 1 and July 1 of that year
usa2010s.drop(['2010'], axis=0, inplace=True)
usa2010s
```

## Data for the 2000s

```python
pep2000s=pd.read_csv(os.path.join(data_path,file2000s),converters={'STATE':str,'COUNTY':str},encoding='LATIN-1')
pep2000s.head()
```

```python
# Pull out records for states and columns for international migration
codes=['STATE','COUNTY']
names=[c for c in pep2000s.columns if c.endswith('NAME')]
net_int=[c for c in pep2000s.columns if c.startswith('INTERNATIONALMIG')]
keep_cols=codes+names+net_int
states2000s=pep2000s.loc[pep2000s['COUNTY']=='000',keep_cols]
states2000s.columns = states2000s.columns.str.replace('INTERNATIONALMIG','')
states2000s.head()
```

```python
states2000s.info()
```

```python
usa2000s=states2000s.sum(axis=0,numeric_only=True)
usa2000s
```

```python
# Drop 2000, as this measures migration between Apr 1 and July 1 of that year
usa2000s.drop(['2000'], axis=0, inplace=True)
usa2000s
```

## Concatenate Data and Plot

```python
us_total=pd.concat([usa2000s,usa2010s])
us_total
```

```python
ax=us_total.plot(kind='bar',figsize=(12,8), legend=False)
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
ax.set_title('US Net International Migration',fontsize=18)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.xticks(rotation=0)
plt.xlabel('\nAnnual estimates as of July 1st, measured from the preceding year\nSource: US Census Bureau 2010 evalaution estimates (2001 to 2010), 2019 vintage population estimates (2011 to 2019)',
          fontsize=12)
ax.set_ylabel("Net Migrants", fontsize=12)
plt.tight_layout()
plt.savefig(os.path.join(out_path, 'images','extras','us_intl_migration.png'),bbox_inches='tight')
```
